## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

A preliminary look at the starter dataset reveals significant variance in page performance and behavior, justifying the need for multi-dimensional clustering. For instance, we can identify pages that are highly visible but fail to capture clicks, alongside pages that receive very few clicks but hold user attention exceptionally well. Simple filters would miss these nuances.

```python
import pandas as pd


df = pd.read_csv('/content/content_refresh_anonymized.csv')

high_imp_low_ctr = df[(df['impressions_90d'] > 1000) & (df['ctr'] < 0.01)]

hidden_gems = df[(df['clicks_90d'] < 50) & (df['engagement_rate'] > 0.60)]

age_spread = df['content_age_days'].describe()

print(f"Total pages analyzed: {len(df)}")
print(f"Pages with >1000 impressions but <1% CTR: {len(high_imp_low_ctr)}")
print(f"Pages with <50 clicks but >60% engagement: {len(hidden_gems)}")
print(f"Content age ranges broadly from {age_spread['min']:.0f} to {age_spread['max']:.0f} days.")
```
